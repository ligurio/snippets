## Мужчины

### Призеры

```python
mens = json.load(open('data/20170813_luzhniki_hm_m_21_1km.json')) # все зарегистрированные
male = [runner for runner in mens['data'] if runner[9] != "DQ"] # только финишеры
print repr([(runner[1], runner[2], runner[3], runner[4]) for runner in male[:5]]).decode('unicode_escape')
```

### Лучший и худший результат

```python
print "Лучшее время у мужчин: ", male[0][9]
print "Худшее время у мужчин: ", male[-1][9]
print "Всего финишировавших мужчин: ", len(male)
```

### Рейтинг беговых клубов по количеству представленных мужчин

```python
def normalize_team(df):
    df.replace([u'I LOVE RUNNING', u'I LOVE RUNNING Мытищи', u'ILoveRunning', u'I love Running', u'I love running', u'I Love Running', u'ILR', u'ILOVERUNNING', u'ILOVERUNNING', u'iloverunning', u'Iloverunning'], u'I Love Running', inplace=True)
    df.replace([u'Academy marathon', u'Academy of marathon', u'Academy Marathon', u'Академия марафона', u'Академия Марафонского Бега'], u'Академия Марафона', inplace=True) 
    df.replace([u'ЦиклON', u'Циклон'], u'ЦиклОн', inplace=True)
    df.replace([u'Gepard', u'GEPARD', u'marathonclubGEPARD'], u'GEPARD', inplace=True)
    df.replace([u'RunLab/AFRC', u'RUNLABclub', u'RUNLAB CLUB', u'RunLab Club'], u'RunLab Club', inplace=True)
    df.replace([u'adidas runners', u'adidas runners AR Нагорныйсила', u'Adidas Runners', u'Adidas', u'ADIDAS RUNNERS', u'Adidas RunClub', u'ADIDAS RUNNERS MOSCOW', u'Адидас', u'Adidas runclub', u'Adidas Runners Moscow', u'Adidas Runclub Fili', u'Adidas Runners Ivanovo', u'Adidas runners', u'Adidas runnes club Mitino', u'Adidas RunClub Sokolniki', u'Adidas Run Club Izmailovo', u'Adidas run club Kirov', u'runess club adidas тушино', u'Adidas Runners Университет', u'Adidas Runners Atrium', u'ADIDAS RUNNERS ULTRA', u'adidas Runners', u'ADIDAS RUNCLUB> ЛОСИНЫЙ', u'AR Ekaterinburg'], u'Adidas Runners', inplace=True) 
    df.replace([u'NIKE RUNNING CLUB', u'Беговой клуб NIKE+', u'Nike+', u'Nike+ Run Club', u'No Tag Runners / Nike+', u'Nike', u'Nike+', u'Nike Running Club', u'Nike running club', u'Nike+ Running', u'Беговой клуб Nike парк Горького', u'Nike Run Club', u'nike+run club', u'Nike Running Russia', u'Nike+ Парк Горького', u'NIKE+'], u'Nike Running', inplace=True)
    df.replace([u'Trilife', u'Трилайф', u'trilife', u'TriLife.ru', u'TriLife'], u'Trilife', inplace=True)
    df.replace([u'Piranha running club', u'Piranha', u'piranha', u'PIRANHA'], u'Piranha', inplace=True)
    df.replace([u'RRUNS', u'Rruns'], u'RRUNS', inplace=True)
    df.replace([u'Нет', u'Лично', u'лично', u'нет', u'0'], u'Личный зачет', inplace=True)
    df.replace([u'БиМ', u'БИМ', u'Бег и мы'], u'БиМ', inplace=True)
    df.replace([u'КЛБ Столица', u'Столица'], u'Столица', inplace=True)
    df.replace([u'RC Лосиный', u'Runclub Лосиный', u'RunClub Лосиный'], u'RC Лосиный', inplace=True)
    df.replace([u'RunLab/AFRC', u'RUNLABclub', u'RUNLAB CLUB', u'RunLab Club'], u'RunLab Club', inplace=True)

male_teams = lambda: (([runner[8], runner[7]]) for runner in male)
dfmt = pd.DataFrame(male_teams(), columns=['team', 'city'])
normalize_team(dfmt)
result = dfmt.groupby('team').aggregate(len)
#result = dfmt.groupby('team').size()
result.sort_values(result.columns[0], ascending=[1], inplace=True)
#result.loc[result['city'] < 100]
result.tail(1000)
#res.plot.bar(figsize=(16, 8), stacked=False);
# игнорировать клубы: нет, Нет, 0, Лично, лично
```

### Рейтинг беговых клубов по количеству мужчин из первой сотни финишеров

```python
rating = dfmt.head(100)
rating100 = rating.groupby('team').aggregate(len)
rating100.sort_values(rating100.columns[0], ascending=[1], inplace=True)
rating100
# игнорировать команды: нет, Нет, 0, Лично, лично
```

### Рейтинг городов по количеству мужчин

```python
male_city = lambda: (([runner[8], runner[7]]) for runner in male)
dfmc = pd.DataFrame(male_city(), columns=['team', 'city'])
cities = dfmc.groupby('city').aggregate(len)
cities.sort_values(cities.columns[0], ascending=True, inplace=True)
cities.tail(20)
```

### Рейтинг стран по количеству мужчин

```python
male_country = lambda: (([runner[8], runner[6]]) for runner in male)

dfmco = pd.DataFrame(male_country(), columns=['team', 'country'])
countries = dfmco.groupby('country').aggregate(len)
countries.sort_values(countries.columns[0], ascending=[1], inplace=True)
countries.tail(9) # все страны с более, чем одним участником
```

### Распределение мужчин по возрасту

```python
male_age = lambda: (([runner[5], runner[6]]) for runner in male)
dfma = pd.DataFrame(male_age(), columns=['age', 'num'])
dfma = dfma.groupby('age').aggregate(len)
dfma.plot.bar(figsize=(16, 8), stacked=False, color='Green', legend=False, title=u"Распределение мужчин по возрасту");
```

### Распределение количества мужчин по времени преодоления дистанции и отдельных сплитов (5, 10, 15 и 20 км)

```python
# https://stackoverflow.com/questions/23966152/how-to-create-a-group-id-based-on-5-minutes-interval-in-pandas-timeseries
def timeseries(data, index):
    time = pd.DataFrame()
    time['time'] = [runner[index] for runner in data]
    time['time'] = pd.to_datetime(time['time'])
    time = time.set_index(['time'])
    time['num'] = [1 for runner in data]
    #time.resample('5Min').sum()
    return time.groupby(pd.TimeGrouper('5Min'))['num'].sum()

#result = pd.concat([timeseries(male, 9), timeseries(male, 11), timeseries(male, 12), timeseries(male, 13), timeseries(male, 14)], columns=['xy'], axis=1, ignore_index=True)
#result.plot(kind='area', figsize=(16, 8), stacked=False, colormap="Greens", title=u"Распределение количества мужчин по времени преодоления дистанции и отдельных сплитов (5, 10, 15 и 20 км)");

from matplotlib.pyplot import *
plt.subplots(figsize=(16, 8))
plt.title(u"Распределение количества мужчин по времени преодоления дистанции и отдельных сплитов (5, 10, 15 и 20 км)")
p1, = plot(timeseries(male, 9))
p2, = plot(timeseries(male, 11))
p3, = plot(timeseries(male, 12))
p4, = plot(timeseries(male, 13))
p5, = plot(timeseries(male, 14))
legend([p1, p2, p3, p4, p5], [u"Финиш", u"5 км", u"10 км", u"15 км", u"20 км"])
```

## Общие данные

```python
total = female + male
print "Всего финишировавших участников: ", len(total)

min = 100
max = 0
imin = []
imax = []
index = 0
for runner in total:
    if min >= runner[5]:
        min = runner[5]
        imin = index
        
    if max < runner[5]:
        max = runner[5]
        imax = index
    index = index + 1

print "Самый молодой участник", min, repr(total[imin]).decode('unicode_escape')
print "Самый старый участник", max, repr(total[imax]).decode('unicode_escape')
```
